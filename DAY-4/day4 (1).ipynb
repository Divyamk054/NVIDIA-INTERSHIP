```python
import torch
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# ==========================================================
# 3. XOR Problem Using Neural Network
# ==========================================================
# XOR Dataset
X = [[0, 0],
     [0, 1],
     [1, 0],
     [1, 1]]

y = [0, 1, 1, 0]

# Neural Network
model = MLPClassifier(hidden_layer_sizes=(4,),
                      max_iter=5000,
                      random_state=42)

model.fit(X, y)

print("\n===== 3. XOR PREDICTIONS =====")
print("Predictions:")
print(model.predict(X))


# ==========================================================
# 4. Visualizing a Neural Network
# ==========================================================
plt.figure(figsize=(6,4))

# Input Layer
plt.scatter([1, 1], [1, 2], s=500, label="Input Layer")

# Hidden Layer
plt.scatter([3, 3, 3], [0.5, 1.5, 2.5], s=500, label="Hidden Layer")

# Output Layer
plt.scatter([5], [1.5], s=500, label="Output Layer")

plt.title("Simple Neural Network Layout")
plt.axis('off')
plt.show()


# ==========================================================
# PyTorch Implementation & MNIST Dataset Setup
# ==========================================================

# # 1. System & Device Configuration
print("🖥️ Step 1: Checking device...")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

if torch.cuda.is_available():
    print("GPU Name:", torch.cuda.get_device_name(0))

# # 2. Automatically Download and Parse MNIST
print("🌐 Step 2: Fetching MNIST via torchvision (Self-Healing)...")
transform = transforms.Compose([
    transforms.ToTensor() # Automatically converts to float32 and scales to [0, 1]
])

# This handles everything: download, extraction, and formatting
mnist_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

# # 3. Step 3: Render Visual Plot
print("🎨 Step 3: Rendering visual plot...")
fig, axes = plt.subplots(2, 5, figsize=(10, 4))
for i, ax in enumerate(axes.flatten()):
    # mnist_dataset[i][0] is the image tensor [1, 28, 28]
    img = mnist_dataset[i][0].squeeze().numpy()
    ax.imshow(img, cmap="gray")
    ax.axis("off")
plt.tight_layout()
plt.show()

# # 4. Step 4: Prepare DataLoader
print("📦 Step 4: Preparing Data Loader...")
loader = DataLoader(
    mnist_dataset,
    batch_size=512,
    shuffle=True,
    num_workers=0  # Safe for all operating systems
)

# Test a batch
images_batch, labels_batch = next(iter(loader))

# Move batch to device for training
images_batch = images_batch.to(device)
labels_batch = labels_batch.to(device)

print("\n🎉 SUCCESS!")
print("Batch Images Shape:", images_batch.shape)  # Expected: [512, 1, 28, 28]
print("Batch Labels Shape:", labels_batch.shape)  # Expected: [512]
print("Active Device:", images_batch.device)
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("▶ Step 1: Generating Input Grid in Memory")
# Create 1 single fake grayscale input image matching MNIST dimensions (28x28)
mock_image = np.random.rand(28, 28)

print("▶ Step 2: Running Structural Parameter Calculations")
# Convolution Layer components: 16 filters, 1 input channel, 3x3 kernel size
filters = 16
in_channels = 1
kernel_w, kernel_h = 3, 3

# Mathematical Parameter Formula: (W * H * C_in * C_out) + Biases
conv_weights_count = kernel_w * kernel_h * in_channels * filters
conv_biases_count = filters
total_conv_params = conv_weights_count + conv_biases_count

print(f" - Conv2D Weights Count : {conv_weights_count}")
# Output: 144
print(f" - Conv2D Biases Count  : {conv_biases_count}")
# Output: 16
print(f" - Total Conv2D Parameters: {total_conv_params}")
# Output: 160

# Fully Connected Layer components
# Max pooling cuts the spatial dimensions directly in half (28x28 -> 14x14)
flatten_features = filters * 14 * 14
output_classes = 10
fc_params = (flatten_features * output_classes) + output_classes

print(f" - Flattened Vector Size: {flatten_features}")
# Output: 3136
print(f" - Dense Layer Parameters: {fc_params}")
# Output: 31370
print(f" - Total Model Parameters : {total_conv_params + fc_params}")
# Output: 31530

print("\n▶ Step 3: Rendering Visual Diagram Pipeline...")
# Plot the mock image input array grid directly
plt.figure(figsize=(5, 5))
plt.imshow(mock_image, cmap='plasma')
plt.title(f"MNIST Matrix Input (28x28)\nConv Layer Parameters = {total_conv_params}", fontsize=11)
plt.colorbar(label="Pixel Intensity Values")
plt.axis('off')
plt.show()

print("☕ Execution Successful! Method 2 bypasses all tensor libraries.")
```
