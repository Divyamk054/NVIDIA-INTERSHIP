```python

import os
import time
import psutil
import torch
import warnings

# Suppress warnings and clean up output windows
warnings.filterwarnings('ignore')
os.environ["TOKENIZERS_PARALLELISM"] = "false"

from transformers import pipeline, AutoTokenizer, AutoModel

print("="*60)
print("LIGHTWEIGHT CPU AI WORKSTATION")
print("="*60)

# Force CPU mode since driver is outdated
device_str = "cpu"
device_id = -1
print(f"\nDevice forced: {device_str.upper()} (Bypassing GPU Driver issues)")

# ---------------------------------------------------
# 1. Tokenizer Demo
# ---------------------------------------------------
print("\n[1/7] TOKENIZATION")
tokenizer_gpt = AutoTokenizer.from_pretrained("distilgpt2")
text = "Artificial Intelligence is transforming education and cybersecurity."
tokens = tokenizer_gpt.tokenize(text)
print("Tokens:", tokens)
print("Token Count:", len(tokens))

# ---------------------------------------------------
# 2. Embedding Demo (SWAPPED: Using tiny 'albert' instead of heavy BERT)
# ---------------------------------------------------
print("\n[2/7] EMBEDDINGS (Loading Tiny ALBERT...)")
tokenizer_tiny = AutoTokenizer.from_pretrained("albert-base-v2")
model_tiny = AutoModel.from_pretrained("albert-base-v2").to(device_str)

inputs = tokenizer_tiny(text, return_tensors="pt").to(device_str)
with torch.no_grad():
    outputs = model_tiny(**inputs)
print("Embedding Shape:", outputs.last_hidden_state.shape)

# ---------------------------------------------------
# 3. Text Generation
# ---------------------------------------------------
print("\n[3/7] TEXT GENERATION")
generator = pipeline("text-generation", model="distilgpt2", device=device_id)
result = generator("Machine Learning is", max_new_tokens=15)
print(result[0]["generated_text"])

# ---------------------------------------------------
# 4. Sentiment Analysis
# ---------------------------------------------------
print("\n[4/7] SENTIMENT ANALYSIS")
sentiment = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", device=device_id)
print(f"Review: 'This workshop is amazing.' -> {sentiment('This workshop is amazing.')[0]}")

# ---------------------------------------------------
# 5. Summarization (SWAPPED: Using tiny 'distilbart' instead of giant BART)
# ---------------------------------------------------
print("\n[5/7] SUMMARIZATION (Loading Mini DistilBART...)")
summarizer = pipeline("summarization", model="sshleifer/distilbart-cnn-6-6", device=device_id)
article = "Artificial Intelligence is a branch of computer science focused on creating systems capable of performing tasks."
summary = summarizer(article, max_length=20, min_length=5)
print(summary[0]["summary_text"])

# ---------------------------------------------------
# 6. Translation
# ---------------------------------------------------
print("\n[6/7] TRANSLATION")
translator = pipeline("translation_en_to_fr", model="t5-small", device=device_id)
translated = translator("Artificial Intelligence is transforming education.")
print(translated[0]["translation_text"])

# ---------------------------------------------------
# 7. Throughput Benchmark & Hardware Monitor
# ---------------------------------------------------
print("\n[7/7] BENCHMARK & HARDWARE")
start = time.time()
output = generator("Explain AI.", max_new_tokens=30)
end = time.time()

elapsed_time = end - start
tokens_generated = len(tokenizer_gpt.encode(output[0]["generated_text"]))
process = psutil.Process()
memory_mb = process.memory_info().rss / (1024 ** 2)

print("Execution Time:", round(elapsed_time, 2), "sec")
print("Performance Throughput:", round(tokens_generated / elapsed_time, 2), "tokens/sec")
print("Host RAM Usage:", round(memory_mb, 2), "MB")
print("\nExecution Complete!")
```
